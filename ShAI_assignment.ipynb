## Data Cleaning

```python
Salaries.isnull().sum()
```

```python
S=Salaries.dropna(axis=1,how='all')
#because Notes and Status are empty for all employees we can drop the whole column 
S.BasePay.fillna(0,inplace=True)
S.OvertimePay.fillna(0,inplace=True)
S.OtherPay.fillna(0,inplace=True)
S.Benefits.fillna(0,inplace=True)
#BasePay,OvertimePay,OtherPay and Benefits are partially empty so we will fill the missing values with 0s.
S.isnull().any()
```
